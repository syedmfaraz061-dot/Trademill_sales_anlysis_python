```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\syed mutallib faraz\\Desktop\\CardioGoodFitness.csv")
df
```

```python
# How many models does store have?

df["Product"].nunique()
df["Product"].value_counts().plot(kind='bar')
plt.title("Model Count")
plt.xticks(rotation=0)
plt.show()

```

```python
# Which is most sold Model?
df["Product"].value_counts()
sns.countplot(data=df, x="Product")
plt.title("Most Sold Model")
plt.show()

```

```python
# Are Male customers buying treadmill more than female customers?
df.groupby("Gender")["Product"].count()
sns.countplot(data=df, x="Gender")
plt.title("Gender vs Treadmill Purchase Count")
plt.show()

```

```python
# What is the Income, Age, Education of people buying treadmill.?
df.groupby("Product")[["Income","Age","Education"]].mean()
```

```python
# How many days and miles customer expect to run on treadmill?

df[["Usage","Miles"]].mean()
sns.barplot(data=df, x="Product", y="Miles")
plt.title("Miles Expected per Model")
plt.show()

sns.barplot(data=df, x="Product", y="Usage")
plt.title("Days Expected per Model")
plt.show()

```

```python
# What is the self-rated fitness of customers buying treadmill.?

sns.countplot(data=df, x="Fitness", hue="Product")
plt.title("Self Rated Fitness vs Model")
plt.show()
```

```python
# Is married customer buying Treadmill more than Single customers?
sns.countplot(data=df, x="MaritalStatus")
plt.title("Marital Status vs Purchase Count")
plt.show()
```

```python
# Is there any relation between Income and model.?
sns.boxplot(data=df, x="Product", y="Income")
plt.title("Income vs Model Bought")
plt.show()
```

```python
# Is there any relation between Age and model?
sns.boxplot(data=df, x="Product", y="Age")
plt.title("Age vs Model Bought")
plt.show()
```

```python
# Is there any relation between self-rated fitness and model?
sns.boxplot(data=df, x="Product", y="Fitness")
plt.title("Self-rated Fitness vs Model")
plt.show()
```

```python
# Is there any relation between education and model.?
sns.boxplot(data=df, x="Product", y="Education")
plt.title("Education Level vs Model Bought")
plt.show()
```

```python
#  Does gender have any effect on model customer buy.?
sns.countplot(data=df, x="Product", hue="Gender")
plt.title("Model Purchased by Gender")
plt.show()
```

```python
# Does Martial status has any effect model customer buy.?
sns.countplot(data=df, x="Product", hue="MaritalStatus")
plt.title("Model Purchased by Marital Status")
plt.show()
```

```python
# Are there different age groups buying different models.?
df["AgeGroup"] = pd.cut(df["Age"], bins=[18,30,45,60,80], labels=["18-30","31-45","46-60","60+"])

sns.countplot(data=df, x="AgeGroup", hue="Product")
plt.title("Age Group vs Model Bought")
plt.show()
```

```python
#  Relation between Age, Income and education and model bought?
sns.pairplot(df, vars=["Age","Income","Education"], hue="Product")
plt.show()
```
